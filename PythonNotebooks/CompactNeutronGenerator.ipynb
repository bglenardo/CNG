# The problem


<img src="files/Diagram of Compact Neutron Generator.png">

The idea behind the compact neutron generator is to have a source of neutrons directly next to the detector, and make calibraiton measurements that have different systematics than the usual collimated, single-position DD measurements.  In two-dimensions, the problem would look something like the above picture.

Now, Kareem's idea is to use the fact that neutrons have certain probabilities of scattering out of the detector depending on where they are, and make spectral difference measurements by searching in certain positions in the detector and comparing to the expected spectra. Here, I'll perform an analytical calculation of the expected spectra.

The basic idea is to solve for $d$, the distance that the neutron travels before it exits the xenon. There will be an exponential attenuation factor of $e^{-d/\lambda}$ due to the finite MFP of neutrons in xenon, which will suppress the single scatter energy spectrum at different energies.

# Deriving $d$ in 2 dimensions

## Rotating for simplicity

We begin by re-arranging the problem so that $d$ is just a radius. This is shown in the following figure:
<img src="files/Copy of Diagram of Compact Neutron Generator.png">


The two straight dotted lines are the x- and y-axes. 

Then, the circle is described by the equation:
\begin{equation}
\frac{(x + b)^2}{a^2} + \frac{y^2}{a^2} = 1
\end{equation}

with $x = d\,\, \text{cos}\, \theta'$ and $y = d\,\, \text{sin}\, \theta'$.  


Then, we plug these in and solve for d:
\begin{equation}
\begin{split}
a^2 & = (d \,\, \text{cos}\, \theta' + b)^2 + (d \,\, \text{sin}\, \theta')^2 \\
\\
& = d^2 \, \text{cos}^2\, \theta' + 2 \, b\,d\,\text{cos}\,\theta' + b^2 + d^2 \, \text{sin}^2\, \theta' \\
\\
& = d^2 + 2\,b\,d\,\text{cos}\,\theta' + b^2
\end{split}
\end{equation}


And then subtracting $a^2$ from both sides gives us
\begin{equation}
d^2 + (2\,b\,\text{cos}\,\theta')\,d + (b^2-a^2) = 0
\end{equation}




To get $d$ from this equation, we just use the quadratic formula
\begin{equation}
\begin{split}
d & = \frac{-2\,b\,\text{cos}\,\theta' \pm \sqrt{4\,b^2\,\text{cos}^2\,\theta' - 4\,(b^2 - a^2)}}{2} \\
\\
& = -b\,\text{cos}\,\theta' \pm b\sqrt{\text{cos}^2\,\theta' - \left(1 - \frac{a^2}{b^2}\right)} \\
\\
& = b \left( \sqrt{\left(\text{cos}^2\,\theta' + \frac{a^2}{b^2} - 1\right)}\,\, - \text{cos}\,\theta'\right)
\end{split}
\end{equation}

We've ignored the - solution, since $d$ is a radius and therefore can't be negative. 

To get a little intuition, we can try a particular case. If $a = b$, $\frac{a^2}{b^2} = 1$. Then $d = 0$ everywhere in the right hemisphere, as we would expect.

```python
theta_prime = np.linspace(0.,2 * np.pi,100)
phi = np.linspace(0.,2 * np.pi,100)
Theta = np.pi/2
b = 55
a = 75
d = b * ( np.sqrt( np.cos(theta_prime)**2 + a**2/b**2 - 1 ) - np.cos(theta_prime))
x = d * np.cos(theta_prime)
y = d * np.sin(theta_prime)

plt.plot(x,y,'.r')
plt.show()

alpha = np.arccos( (a + b * np.cos(Theta)) / (np.sqrt( a**2 + b**2 + 2 *a * b * np.cos(Theta) ))  )
d = d = b * ( np.sqrt( np.cos(phi - Theta)**2 + a**2/b**2 - 1 ) - np.cos(phi - Theta))
x = d * np.cos(phi)
y = d * np.sin(phi)

plt.plot(x,y,'.g')
plt.show()
```

## Plugging in for the correct coordinates.

Now looking back at the first diagram, we see that $\theta' = \phi - \alpha$. Here we redraw the triangle for reference.

<img src="files/Triangle of interest.png">

But what is $\alpha$?


The angle between the $a$ and $b$ legs is $\pi - \theta$. Then, from the law of cosines, we have:
\begin{equation}
\begin{split}
c^2 & = a^2 + b^2 - 2\,a\,b\,\text{cos}\,(\pi - \theta) \\
& = a^2 + b^2 + 2\,a\,b\,\text{cos}\,\theta
\end{split}
\end{equation}


Which tells us that 
\begin{equation}
\text{cos}\,\alpha = \frac{a + b\,\text{cos} \,\theta}{\sqrt{a^2 + b^2 + 2\,a\,b\,
\text{cos}\,\theta}}
\end{equation}

## Putting it together

\begin{equation}
d(a,b,\theta,\phi) = a \left( \sqrt{\left(\text{cos}\,(\phi - \theta) + \frac{a^2}{b^2} - 1\right)}\,\, - \text{cos}\,(\phi - \theta)\right)
\end{equation}

with

\begin{equation}
\alpha = \text{cos}^{-1} \left(\frac{a + b\,\text{cos} \,\theta}{\sqrt{a^2 + b^2 + 2\,a\,b\,
\text{cos}\,\theta}}\right)
\end{equation}







#    Generalizing to scattering in 3D

Now, we want to let the neutron travel off the plane of the CNG. So $d$ becomes one of two variables that are important for the distance to escape; the other being $z$.  

For simplicity, we'll approximate the can as an infinite 3D cylinder.  The calculation including the top and bottom seems pretty daunting, as you'd have to write the planar surfaces in curvilinear coordinates. Instead, we can just use a piecewise calculation when doing the spectra; if $z$ for a given scattering angle is too high or low, just fix it at the top or bottom.


For the infinitely long cylinder, the path to escape becomes

\begin{equation}
D^2 = d^2 + d^2 \, \text{tan}^2\,\gamma
\end{equation}

where $\gamma$ is the angle off the plane and $ z = d\,\text{tan}\,\gamma$.

Now the calculation of the actual scattering angle becomes tricky, since we have to combine $\gamma$ with $\theta'$. 

I think we can just do this using the dot product and spherical coordinates. We have a sphere where the x-axis is the axis that the incident neutron coming from, defined by $\alpha$, and the xy-plane is the point at which the polar angle from the zenith is $\pi$.  Then we just take the dot product of the $x$ unit vector ( $\hat{x}$ ) with the vector of coordinates $\hat{s} =$ ( $r = 1$, $\text{zenith} = \pi/2 - \gamma$, $\text{azimuth} = \theta'$ ).

This product is

\begin{equation}
\begin{split}
\hat{x} \cdot \hat{s} = ||\hat{x} ||\, ||\hat{s}|| \,\text{cos}\, \Theta  & = (\, \hat{x}\, ) \cdot (\, \text{sin}\,(\pi/2 - \gamma)\,\text{cos}\,(\theta' + \theta - \alpha)\,\hat{x} +  \text{sin}\,(\pi/2 - \gamma)\,\text{sin}\,(\theta'  + \theta - \alpha)\,\hat{y} + \text{cos}\,(\pi/2 - \gamma)\,\hat{z}\, ) \\
& = \text{sin}\,(\pi/2 - \gamma)\,\text{cos}\,(\theta' + \theta - \alpha) 
\end{split}
\end{equation}

So

\begin{equation}
\Theta = \text{cos}^{-1}\, \left( \text{sin}\,(\pi/2 - \gamma)\,\text{cos}\,(\theta' + \theta - \alpha)  \right)
\end{equation}
